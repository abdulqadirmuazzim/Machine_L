```python
"""
A simple example of an animated plot
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x1 = np.arange(0, 2*np.pi, 0.01)        # x-array
line, = ax.plot(x1, np.sin(x1))
def animate(i):
    line.set_ydata(np.sin(x1+i/10.0))  # update the data
    return line,

#Init only required for blitting to give a clean slate.
def init():
    line.set_ydata(np.ma.array(x1, mask=True))
    return line,

ani = animation.FuncAnimation(plt.gcf(), animate, np.arange(1, 200), init_func=init,
    interval=25, blit=True)
plt.show()
```

```python
hor = list(x)
ver = list(a)
work = plt.plot([])

pen = work[0]
# plt.cla()
# function
def run(f):
    pen.set_xdata(hor)
    pen.set_ydata(ver)
    plt.plot(hor[f], ver[f])
    plt.scatter(hor[f], ver[f], 30, color='black')
#     plt.xlim(1950,2019)
#     plt.ylim(-10, 80)

ani = animation.FuncAnimation(plt.gcf(), func=run, frames=len(hor), interval=25)
plt.show()
```
