```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib notebook

data = pd.read_csv('africa_c.csv')
data
```

```python
x = data['year'][650:708]
y = data['exch_usd'][650:708]
a = data['inflation_annual_cpi'][650:708]

plt.title("Nigerian USD Exchange")
plt.xlabel("year")
plt.ylabel('exch_usd')
plt.scatter(x, y, color='blue')
plt.plot(x, y, color='green')
plt.fill_between(x, y, alpha=0.8,interpolate=True, color='red')
plt.show()
plt.savefig('niger.png')
```

```python
plt.style.use('ggplot')
print(plt.style.available)
plt.title('Nigerian Annual Inflation CPI Graph')
plt.xlabel('Year')
plt.ylabel("inflation_annual_cpi")
plt.scatter(x,a, color='blue')
plt.fill_between(x,a,alpha=0.7, color='black')
plt.plot(x, a , color="orange")
plt.show()
```
