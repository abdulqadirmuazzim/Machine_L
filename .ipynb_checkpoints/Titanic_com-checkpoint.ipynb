```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.tree import DecisionTreeRegressor as dtr
from sklearn.neighbors import KNeighborsClassifier as knc
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')

train = df1.replace(['male','female'], [1,0])
test = df2.replace(['male','female'], [1,0])
train
df1
```

```python

tr =train.drop(columns=['Name','Ticket', 'Cabin','Embarked', 'Survived', "PassengerId", 'Fare'])
te =test.drop(columns=['Name','Ticket', 'Cabin','Embarked', 'PassengerId', 'Fare'])
out = train['Survived']
tr = tr.fillna(-1)
te = te.fillna(-1)
# for i,c in enumerate(zip(out, df1['Sex'])):
#     print(i,f'survived{c}')
print(tr, train['Survived'])
```

```python
from sklearn import metrics
from sklearn import svm
import openpyxl as op

xtr, xte, ytr, yte = tts(tr, out, test_size=.2)

model = svm.SVC()

model.fit(xtr, ytr)
prediction = model.predict(te)

prediction
```

```python
deci = dtc()
deci.fit(xtr,ytr)
pred = deci.predict(te)


pred
```

```python

# sigmoid function
def sig(x):
    return 1/(1+np.exp(-x))

# change in sigmoid
def d_sig(s):
    return sig(s) * (1-sig(s))

# ploting sigmoid & change in sigmoid
x = np.linspace(-5, 5, 100)
y = sig(x)
z = d_sig(x)
plt.plot(x, y, color='blue')
plt.plot(x, z, color="red")
plt.show()
```

```python
# there is going to be 5 weights and bios
weights = np.random.rand(5)
b = np.random.rand()

# main function

def cal(w, inp, b):
    ans = ((w[0] * inp[0]) + (w[1] * inp[1]) + (w[2] * inp[2]) + (w[3] * inp[3]) + (w[4] * inp[4])) + b
    return ans
    
weights[0]  

cal(weights, tr.loc[2], b)

costl = []
print(weights)
# iterations
loops = 10000
learn_r = 0.1
for a in range(loops):
#     random index
    ind = np.random.randint(len(tr))
#     target
    tar = train['Survived'][ind]
    z = cal(weights, tr.loc[ind], b)
    ans = sig(z)
#     cost
    cost = (ans - tar)**2
#     change in cost & sigmoid
    dydsig = d_sig(z) 
    dydc = 2*(ans - tar)
#     changes in ans with respect to weights
    dydw1 = tr.loc[ind][0]
    dydw2 = tr.loc[ind][1]
    dydw3 = tr.loc[ind][2]
    dydw4 = tr.loc[ind][3]
    dydw5 = tr.loc[ind][4]
    dydb = 1
#     updating cost list
    if a % 1000 == 0:
        costl.append(cost)
        print(ans,  tar)
        
    dydx = dydc * dydsig
#     changes in cost with respect to weights
    dcdw1 = dydx * dydw1
    dcdw2 = dydx * dydw2
    dcdw3 = dydx * dydw3
    dcdw4 = dydx * dydw4
    dcdw5 = dydx * dydw5
    dcdb = dydx * dydb
#     update the weights
    weights[0] -= learn_r * dcdw1
    weights[1] -= learn_r * dcdw2
    weights[2] -= learn_r * dcdw3
    weights[3] -= learn_r * dcdw4
    weights[4] -= learn_r * dcdw5
    b -= learn_r * dcdb

print(weights)    
plt.plot(range(len(costl)), costl)
plt.show()   
```
