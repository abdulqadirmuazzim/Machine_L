```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

data = pd.read_csv('i-p.csv', index_col='Year')
data = data.fillna(0)
data
```

```python
def clean(arg):
    
    arg = str(arg)
    
    if 'billion' in arg:
        ans = arg.strip('billion')
        ans = float(ans)
        return ans * 1e9
    
    if 'million' in arg:
        ans = arg.strip('million')
        ans = float(ans)
        return ans* 1e6
    
    if ',' in  arg:
        com = arg.replace(',', '')
        return float(com)
    
    else:
        return float(arg)


ff = '222,000,000,'

for a in data.drop('Country', axis=1):
    data[a] =  data[a].apply(clean)

data.tail()
```

```python
high_n = ['GDP (in USD)', 'Population', 'Agricultural Output (in USD)', 'Active Military Personnel', 'Reserve Military Personnel', 'IT Output (in USD)', 'Number of Tanks', 'Number of Armoured Vehicles']

isl, pls = data.groupby('Country')

isl = np.array(isl, dtype=object)

new_is = isl[1].drop('Country', axis=1)

pls = np.array(pls, dtype=object)

new_pl = pls[1].drop('Country', axis=1)
```

```python
plt.figure(figsize=(19, 10))

# plot = new_is[ 'GDP (in USD)'].join(new_pl[ 'GDP (in USD)'], lsuffix='_Iseral', rsuffix='_Palestine')
# plot = new_pl.join(new_is, lsuffix='_pl', rsuffix='_is')

sb.heatmap(data=new_pl.drop(columns=high_n), annot=True)

plt.show()
```
