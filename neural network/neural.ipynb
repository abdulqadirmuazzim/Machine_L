```python
# equation = x^2 - 8x + 16 = 0
x_data = []
y_data = []
# getting the values of y
for x in range(-50, 50):
    soln = (x**2) - (8*x) + 16
    x_data.append(x)
    y_data.append(soln)

plt.title("Quadratic curve")
plt.xlabel('Ranges of x')
plt.ylabel('Corresponding values of y')
plt.xlim(-100, 100)
# plt.scatter(x_data, y_data, 10, color='coral')
plt.plot(x_data, y_data, color='darkblue')
plt.plot(x_data, [0 for a in x_data], color='black')
plt.plot([0 for a in x_data], y_data, color='black')
plt.show()
```
